변수 선언
x = [1 2 3], y = [1 3 2]

```python
x = [1, 2, 3]
y = [1, 3, 2] 
```

xbar, ybar 계산

```python
n = 3

xbar = sum(x)/n
ybar = sum(y)/n

xbar,ybar
```

image.png

```python
sumxx=0
sumxy=0
for i in range(n):
    sumxx +=(x[i] - xbar)**2
    sumxy +=(x[i] - xbar)*(y[i]-ybar)
```

alpha와 beta 구하기

image.png

```python
beta = sumxy/sumxx

alpha = ybar - beta * xbar

alpha, beta
```

이후 그래프를 구하면 다음과 같다.

최소자승법 결과 구해진 그래프는 과제와 다르게 나왔다.

이는 검산 결과 sumxx=2.0, sumxy=1.0이 나오면서 beta가 0.5가 되었기 때문이었다.

이를 통해  최소자승법을 통해 구한 간단한 선형 회귀 그래프가 예상과는 다름을 알 수 있었다.


**H**.W #**2** 다변수함수 선형회귀

행렬과 전치행렬 선언

image.png

```python
import numpy as np

#  X, Y와 전치행렬 XT를 선언
X = np.array([[1, 1], [1,2], [1, 3]])
Y = np.array([1, 3, 2])
XT = np.transpose(X)
```

image.png

```python
A=np.linalg.inv(np.dot(XT,X))
B = np.dot(XT, Y)
```

image.png

```python
beta = np.dot(A,B)
```

```python
print(beta)
```
