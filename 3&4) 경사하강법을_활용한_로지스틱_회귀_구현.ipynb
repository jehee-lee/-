X행렬과 Y행렬을 정의하고, beta값인 b와 w의 초기값을 임의의 값(5,5)으로 설정한다.

```python
import numpy as np

X=np.array([1, 2, 3, 4, 5, 6])
Y=np.array([0,0,0,1,1,1])

b=5
w=5

beta= np.array([b,w])
```

시그모이드 함수

```python
def fun_S(w,x,b):
  return 1/(1+np.exp(-w*x-b))
```

H.W#3에서의 결과에 따라 b,w로 각각 편미분한 Cost값을 구한다.

이 때

학습률을 0.1로 설정하고

학습된 b,w를 구함

```python
b_cost_list=[]
b_list=[]
w_cost_list=[]
w_list=[]

#2000번 계산함
for epoch in range(2000):
  dif_b_cost=0
  dif_w_cost=0

#X,Y성분값을 Summary한 값을 구함
  for i in range(6):
    dif_b_cost+=-(Y[i]-fun_S(w,X[i],b))
    dif_w_cost+=-(Y[i]-fun_S(w,X[i],b))*X[i]

  b_cost_list.append(dif_b_cost)
  w_cost_list.append(dif_w_cost)

  #학습률을 0.1로 하고 상기 b,w값을 갱신함)
  lam=0.1

  b=b-lam*dif_b_cost
  w=w-lam*dif_w_cost

  b_list.append(b)
  w_list.append(w)

  beta1= np.array([b,w])
  if epoch %100 == 0:
      print('dif_b_cost=%f, dif_w_cost=%f' %(dif_b_cost, dif_w_cost),'    ||  beta=', beta1)
```

해당 값을 그래프로 plot

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2,2)

fig.set_figheight(5)
fig.set_figwidth(5)

ax[0, 0].plot(b_cost_list)
ax[0, 1].plot(w_cost_list)
ax[1, 0].plot(b_list)
ax[1, 1].plot(w_list)

ax[0, 0].set_title("cost function of b")
ax[0, 1].set_title("cost function of w")
ax[1, 0].set_title("b")
ax[1, 1].set_title("w")

fig.tight_layout()
plt.show()
```
